## 1. Relationship plots

### 1. lineplot

```python
# Use this line to change the widht and height of the seaborn plot.
plt.figure(figsize = (10, 6))

# Display pixel per inch, imporves the visibility of the plot.
plt.figure(dpi = 150)

# Set the title of the plot.
plt.title("Relationship between EngagementSurvey and Salary")

sns.lineplot(x = "EngagementSurvey", y = "Salary", data = human_resources_data)
```

#### Some examples of lineplot.

```python
plt.figure(figsize = (10, 6), dpi = 150)
plt.title("Relationship between SpecialProjjectsCount and Salary")
sns.lineplot(x = "SpecialProjectsCount", y = "Salary", data = human_resources_data)
```

## 2. Categorical plots

### 1. countplot

```python
plt.figure(figsize = (18, 6))

# Set the title of the plot.
plt.title("Number of Employees hired from each Platforms.")

sns.countplot(x = "RecruitmentSource", data = human_resources_data)
```

#### Some examples of countplot

```python
plt.figure(figsize = (8, 4), dpi = 150)

# setting the different color palette
color_palette = sns.color_palette("plasma")
sns.set_palette(color_palette)

sns.countplot(x = human_resources_data["PerformanceScore"], data = human_resources_data)
```

```python
plt.figure(figsize = (8, 4), dpi = 150)

sns.countplot(x = human_resources_data["MarriedID"], data = human_resources_data)
```

### 2. boxplot

```python
plt.figure(figsize = (10, 6), dpi = 100)
sns.boxplot(x = "Category", y = "Sales",  data = super_store_data)
```

### 3. violinplot

```python
plt.figure(figsize = (10, 6), dpi = 100)

# setting the different color palette
color_palette = sns.color_palette("Accent_r")
sns.set_palette(color_palette)

sns.violinplot(x = "Category", y = "Sales",  data = super_store_data)
```

### Pro tip: 
You can use sns.catplot() it provides a kind perameter in which you can specify the type of plot you want. Possible values are, violin, box, bar, count, swarm.

#### Some examples of catplot examples

```python
plt.figure(figsize = (10, 6), dpi = 100)
sns.catplot(x = "Category", y = "Sales",  data = super_store_data, kind = "violin")
```

```python
plt.figure(figsize = (10, 6), dpi = 100)
sns.catplot(x = "Category", y = "Sales",  data = super_store_data, kind = "bar")
```

## 3. Distribution plots

### 1. displot

```python
plt.figure(figsize = (10, 6), dpi = 200)
sns.displot(np.log1p(super_store_data['Sales']))
```

### 2. kdeplot

```python
plt.figure(figsize = (10, 6), dpi = 150)
sns.kdeplot(data = super_store_data, x = "Sales", shade = True)
```

```python
plt.figure(figsize = (10, 6), dpi = 150)
sns.kdeplot(data = super_store_data, x = "Profit",shade = True)
```

## 4. Matrix plots

```python
plt.figure(figsize = (10, 6), dpi = 150)

# setting the color palette
color = sns.color_palette("BuGn_r")
sns.set_palette(color)

corr_matrix = super_store_data.corr()

plt.figure(figsize = (10, 5))
sns.heatmap(corr_matrix, vmax=0.9, annot=True, cmap=color)
```
